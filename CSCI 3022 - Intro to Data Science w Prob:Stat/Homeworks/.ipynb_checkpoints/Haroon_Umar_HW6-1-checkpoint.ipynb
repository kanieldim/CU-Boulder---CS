## [20 points] Problem 1 - Random Samples from ANYWHERE


In notebook 12, we looked at one method many pieces of statistical software use to turn pseudorandom numbers coming from the $U(0,1)$ distribution into those with a normal distribution.  In this problem we examine another such method.

### a) Simulating an Exponential
#### i ) The exponential distribution has pdf $f(x)= \lambda e^{-\lambda x}$ for $x\ge 0$.  Use the following markdown cell to compute by hand the cdf of the exponential.

**Integrate the PDF**

$$\int_0^x\lambda e^{-\lambda x}$$
$$\lambda\int_0^x e^{-\lambda x}$$
$$ -e^{-\lambda x}|_0^x$$
$$ -e^{-\lambda x}+e^{0}$$

The CDF is $ F(x)=-e^{-\lambda x}+1$ for $x\ge 0$

#### ii) The cdf is a function that takes x-values (or times, in the case of the exponential) and returns probabilities as the y-values.  Specifically, it returns the probability $P(X \leq x)$.  Find the *inverse* of this function, that takes as input probabilities and outputs times.

We want the inverse, $F^{-1}(y)$
<br>
We already have $1 - e^{-\lambda x}$
<br>
So, if we set the inverse equal to this, we can easily solve:
$$ y =  1 - e^{-\lambda x} $$
$$ e^{-\lambda x} = 1 - y $$
$$ -\lambda x = ln(1 - y) $$
$$ x = \frac{ln(1 - y)} {-\lambda}$$


#### iii) Simulate 1000 random $U(0,1)$ variables.  Since these are numbers in $[0,1]$, we could think of them as random *probabilities*.  Plug them into the function you found in ii), then plot a histogram of the results.  Overlay the theoretical density of the exponential.  For each of these, use $\lambda=1/4$.

```python
array = []
def inverse():
    for i in range(1000):
        x = random.uniform(0, 1)
        y = -4 * (np.log(1 - x))
        array.append(y)
    plt.hist(array, range=[0,20])
    plt.xlabel('Inverse Function')
    plt.ylabel('Frequency')
    plt.show()
    
inverse()
```

### b) Simulating a new distribution
#### i) Consider a continuous random variable given by $f(x)=cos(x)$ for $x \in [0, \pi/2]$.  Find the cdf and inverse cdf of X.

**Integrate the PDF**

$$\int_0^{\frac{\pi}{2}}cos(x)$$
$$ sin(x)|_0^{\frac{\pi}{2}}$$


The CDF is $ F(x)=sin(x)$ for $x \in [0, \pi/2]$

the inverse cdf is $F^{-1}(x) = sin^{-1}(x)$


#### ii) Simulate draws from X by simulating 1000 $U[0,1]$ random variables and plugging them into the inverse cdf in i).  As in part a), plot a histogram of the random variables against their density function.

```python
array = []
def inverse():
    for i in range(1000):
        x = random.uniform(0, 1)
        y = math.asin(x)
        array.append(y)
    plt.hist(array, range=[0, math.pi / 2])
    plt.xlabel('Inverse Function')
    plt.ylabel('Frequency')
    plt.show()
    
inverse()
```

### c) Backwards Normals

The process we're following could be summarized as follows, referencing functions we know for many distributions:

- Sample from U(0,1)
- Plug those samples each into the .ppf ("Percentile Point Function," or inverse of the CDF) for the random variable.
- Save the output

Now, follow this process for normal random variables.  Sample 1000 random normal variables using `norm.rvs` and then generate 100 random normal variable by transforming 1000 random variables from their `norm.ppfs`.  Plot both samples in an overlapping histogram.

```python
array1 = []
array2 = []
def normals():
    randomize = np.random.uniform(0,1,1000)
    rvs = stats.norm.rvs(0,1,1000)
    array1.append(stats.norm.ppf(randomize))
    array2.append(stats.norm.ppf(rvs))
    plt.hist(array1, alpha = 0.5, density = True)
    plt.hist(array2, alpha = 0.5, density = True)
    plt.ylabel('Frequency')
    plt.xlabel('Inverse Function')
    plt.show()
normals()
```

```python
# GIVEN CODE:
def confidence_intervals(m=500, n=100):
    #get a z value here, the normal way wasn't providing the correct z
    z = NormalDist().inv_cdf((1 + 0.66) / 2)
    truevar = 400
    truestandard = 20
    trueMean = 40
    k = 4
    theta = 10
    n_sample = 100
    sampleArray = []
    meanArray = []
    upperBound = []
    lowerBound = []
    countNumCI = 0
    #over the 500 times calculate the rvs and CI
    for i in range(0, 499):
        x = stats.gamma.rvs(k, loc = 0, scale = theta, size = n, random_state=None)
        sampleArray.append(x)
        meanArray.append(np.mean(x))
        upperBound.append(np.mean(x) + (z * (2)))
        lowerBound.append(np.mean(x) - (z * (2)))
    # increment ci count if mean in CI
    for i in range(0, 499):
        if((upperBound[i] >= trueMean) and (lowerBound[i] <= trueMean)):
           countNumCI += 1
    ## for plotting peurposes look at the ci bounds and mean 
    for i in range(0, 49):
        sample = random.randint(0, 499)
        if ((upperBound[i] >= trueMean) and (lowerBound[i] <= trueMean)):
            plt.vlines(i, lowerBound[i], upperBound[i], linestyles = "solid", colors = "steelblue") 
        else:
            plt.vlines(i, lowerBound[i], upperBound[i], linestyles ="solid", colors ="coral") 
    plt.axhline(y = trueMean, color = "black", linestyle = '-', label = "True Mean")
    plt.legend()

    plt.show()
    print(f"The proportion of confidence intervals that successfully cover the true mean of the distribution is {countNumCI}/500 which is around 66%")
confidence_intervals()
```

**Part D:** Your mysterious employer owns a lot of beachfront property in Hawaii and does not want to see people shy away from buying real estate and spending time near the sea. So, they really want to convince others to that the mean sea level in Honolulu in not increasing.
```python
#use helper funciotns for my own sanity
def getDataForDecade(firstYear):
    return cleaned[cleaned["year"].between(firstYear, firstYear + 10)]["seaLevel"]
#calculate CI and put into list
def getCIValue(data):
    lower = np.average(data) - (stats.norm.ppf(0.99))*(np.std(data)/math.sqrt(len(data)))
    higher = np.average(data) + (stats.norm.ppf(0.99))*(np.std(data)/math.sqrt(len(data)))
    return [lower, higher]

    
years = []
averagesForDecade = []
expectedDecadeCI = getCIValue(cleaned[cleaned["year"].between(2000, 2010)]["seaLevel"])
#get all averaged data
for year in range(cleaned.iloc[0].year, 2000):
    averagesForDecade.append(np.average(getDataForDecade(year)))
    years.append(year)
# get which indexes in DF is within CI
indicesOfInterest = []
for index, avg in enumerate(averagesForDecade):
    if avg >= expectedDecadeCI[0] and avg <= expectedDecadeCI[1]:
        indicesOfInterest.append(index)
#sort indices just in case
indicesOfInterest.sort()
#year at index to get the start of decade
firstYear = years[indicesOfInterest[0]]
#add 10 for decade
secondYear = firstYear + 10
print(f"with 99% certainty, the decade {firstYear} to {secondYear} is statistically within the range of the decade 2000-2010")
```
